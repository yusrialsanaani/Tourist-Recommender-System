#DATAFRAME 

```python
data=pd.read_csv('/content/drive/MyDrive/travel_rs/attraction_info_df_final_1.csv')
data.head()
```

#SCTTER PLOT THAT SHOWS ALL THE ATTRACTIONS ACROSS DIFFERNT PROVIENCES WITH ITS RATINGS


```python
  

fig = go.Figure(go.Scatter(
    y=data["provience"],
    x=data["no. of rating"],
    text=data["location_name"],
    mode='markers',
    
    
  
))
# Tune marker appearance and layout
fig.update_traces(mode='markers', marker=dict(sizemode='area'
                                             , line_width=2))


fig.update_layout(
    title='attractions per provience',
    xaxis=dict(
        title='user rating',
        gridcolor='white',
        type='log',
        gridwidth=2,
    ),
    yaxis=dict(
        title='proviences in canada',
        gridcolor='white',
        gridwidth=2,
    )
    
)

fig.show()
```

#SCATTER PLOT FOR TOURIST ATTRACTIONS HAVING MORE THAN 10K REVIEWS


```python
data["rat"] = data["no. of rating"].str.replace(",","").astype(int)
d12=data[data.rat>10000]
d12
```

```python
  

fig = go.Figure(go.Scatter(
    y=d12["provience"],
    x=(d12['rat']),
    text=d12["location_name"],
    mode='markers',
    
    
  
))
# Tune marker appearance and layout
fig.update_traces(mode='markers', marker=dict(sizemode='area'
                                             , line_width=2))


fig.update_layout(
    title='attractions list with ratings more than 10k',
    xaxis=dict(
        title='user rating',
        gridcolor='white',
        type='log',
        gridwidth=2,
    ),
    yaxis=dict(
        title='proviences in canada',
        gridcolor='white',
        gridwidth=2,
    )
    
)

fig.show()
```
